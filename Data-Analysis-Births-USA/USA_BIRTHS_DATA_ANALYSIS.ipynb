```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("./US_births_2000-2014.csv")
np.sum(dataset.births)
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
# Total Number of births per Year
year2000 = np.sum(dataset.births[dataset.year==2000])
year2001 = np.sum(dataset.births[dataset.year==2001])
year2002 = np.sum(dataset.births[dataset.year==2002])
year2003 = np.sum(dataset.births[dataset.year==2003])
year2004 = np.sum(dataset.births[dataset.year==2004])
year2005 = np.sum(dataset.births[dataset.year==2005])
year2006 = np.sum(dataset.births[dataset.year==2006])
year2007 = np.sum(dataset.births[dataset.year==2007])
year2008 = np.sum(dataset.births[dataset.year==2008])
year2009 = np.sum(dataset.births[dataset.year==2009])
year2010 = np.sum(dataset.births[dataset.year==2010])
year2011 = np.sum(dataset.births[dataset.year==2011])
year2012 = np.sum(dataset.births[dataset.year==2012])
year2013 = np.sum(dataset.births[dataset.year==2013])
year2014 = np.sum(dataset.births[dataset.year==2014])

rownames = ("year2000","year2001","year2002","year2003","year2004","year2005","year2006","year2007","year2008","year2009","year2010","year2011","year2012","year2013","year2014")
# Cumulative Total of births Per Year Dataframe
TotalYear = np.array([year2000,year2001,year2002,year2003,year2004,year2005,year2006,year2007,year2008,year2009,year2010,year2011,year2012,year2013,year2014])
TotalPerYear = pd.DataFrame([year2000,year2001,year2002,year2003,year2004,year2005,year2006,year2007,year2008,year2009,year2010,year2011,year2012,year2013,year2014],index=rownames)
np.max(TotalPerYear),TotalPerYear
```

```python
# BarPlot Showing Distribution Of Births Over The Years
scale = TotalYear/10000
y_pos=np.arange(len(rownames))
plt.figure(figsize=(15,10))
plt.barh(y_pos,scale,0.6,align='center',color='g',alpha=0.8)
plt.yticks(y_pos,rownames)
plt.ylabel("Years Of Birth")
plt.title("Births In USA Between 2000-2014")
```

```python
# Total Per Month
Jan = np.sum(dataset.births[dataset.month==1])
Feb = np.sum(dataset.births[dataset.month==2])
Mar = np.sum(dataset.births[dataset.month==3])
Apr = np.sum(dataset.births[dataset.month==4])
May = np.sum(dataset.births[dataset.month==5])
Jun = np.sum(dataset.births[dataset.month==6])
Jul = np.sum(dataset.births[dataset.month==7])
Aug = np.sum(dataset.births[dataset.month==8])
Sept = np.sum(dataset.births[dataset.month==9])
Oct = np.sum(dataset.births[dataset.month==10])
Nov = np.sum(dataset.births[dataset.month==11])
Dec = np.sum(dataset.births[dataset.month==12])

barNames =("Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sept","Oct","Nov","Dec")

TotalMonth = np.array([Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sept,Oct,Nov,Dec])
TotalPerMonth = pd.DataFrame([Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sept,Oct,Nov,Dec],barNames)
```

```python
# barChart Showing The Population Per Month In the Years
y_pos = np.arange(len(barNames))
plt.figure(figsize=(15,8))
plt.bar(y_pos,TotalMonth,0.6,color='bg',alpha=0.7)
plt.xticks(y_pos,barNames)
plt.ylabel("Birth Per Month",fontSize=20)
plt.title("USA BIRTHS PER MONTH BETWEEN 2000-2014",fontSize=40)
plt.show()
```

```python
# Line Graph Showing The Distribution Of 2000
y_pos = np.arange(len(rownames))
plt.figure(figsize=(20,5))
plt.plot(Births_Per_Month_2000,color='g')
plt.scatter(y_pos,Births_Per_Month_2000,color='r')
plt.xticks(y_pos,rownames,fontSize=15)
plt.xlabel("Months",fontSize=30)
plt.ylabel("Births Per Month",fontSize=30)
plt.title("Births In 2000 Per Month",fontSize=50,color='b',alpha=.7)
plt.show()
```
